The former employeesâ€™ projects get delayed, which makes it difficult to meet timelines, resulting in a reputation loss among consumers and partners
A sizeable department has to be maintained, for the purposes of recruiting new talent
More often than not, the new employees have to be trained for the job and/or given time to acclimatise themselves to the company
Hence, the management has contracted an HR analytics firm to understand what factors they should focus on, in order to curb attrition. In other words, they want to know what changes they should make to their workplace, in order to get most of their employees to stay. Also, they want to know which of these variables is most important and needs to be addressed right away.

# Room For Improvement

Based on EDA above, employee who has higher risk to leave the company are those who have the following criteria:
1. Age: Employee in age range 21-30 are at higher risk to leave the company
2. Marital Status: Employee who are still single are at higher risk to leave the company
3. Business Travel: Emplpoyee with higher travel frequency are at higher risk to leave the company
4. Environment Satisfaction, Work Life Balance, and Job Satisfaction: The lower the value an employee got, the higher is the risk to leave the company
5. Years Since Last Promotion: Employee who didn't get promotion more than 5 years are at higher risk to leave the company
6. Years With Current Manager: More the time spend with their current manager, more less the risk to leave the company

If the Company (Especially HR team) want to keep their current employee, they need to be aware for the cause why the employee left the company and can make a prevention move to keep their employee (such as not making an employee has more business travel time than the others)
